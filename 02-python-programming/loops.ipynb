```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

nested loop <br>
loop inside a loop

```python
for i in range(3):
    for j in range(2):
        print(f"i is {i}, and j is {j}")
```

calculate the sum of first N natural numbers using while loop

```python
n = 1
sum = 0
while n<=10:
    sum=sum+n
    n+=1
print(sum)
```

calculate the sum of first N natural numbers using for loop

```python
sum = 0
for i in range(1, 11):
    sum = sum+i
print(sum)
```

print prime numbers between 1 to 100

```python
for num in range(1, 100):
    if num>1:
        for i in range(2, num):
            if num%i==0:
                break
        else:
            print(num)
```
