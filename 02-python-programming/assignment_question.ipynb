**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
num = float(input("enter the number"))

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(num))
```
